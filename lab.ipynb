```python
# plot some noise audio clips and some data audio clips to look att the data
import numpy as np
import matplotlib.pyplot as plt
import os
import glob
import librosa
```

```python
data_audio_clips = glob.glob("data/speech_commands_v2/*/*.wav")

# read the first file
data_audio_clip = data_audio_clips[2]
data_audio, sr = librosa.load(data_audio_clip, sr=None)

# plot the audio file
plt.figure()
plt.plot(data_audio)
plt.title("Data audio clip")
plt.show()
```

```python
# do the same for noise

noise_audio_clips = glob.glob("data/urban_sound/*.wav")

# read the first file
noise_audio_clip = noise_audio_clips[10]
noise_audio, sr = librosa.load(noise_audio_clip, sr=None)

# plot the audio file
plt.figure()
plt.plot(noise_audio)
plt.title("Noise audio clip")
plt.show()
```
